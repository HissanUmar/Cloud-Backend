```python
import os
from google.colab import userdata

os.environ["HUGGINGFACEHUB_API_TOKEN"] = userdata.get("HF_TOKEN")
os.environ["TAVILY_API_KEY"] = userdata.get("TAVILY_API_KEY")
```

```python
from crewai import Agent, Task, Crew
from crewai_tools import TavilySearchTool
# from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
from crewai import LLM

# llm = HuggingFaceEndpoint(
#     repo_id="Qwen/Qwen2.5-7B-Instruct",
#     provider="auto",
#     task="text-generation",
#     max_new_tokens=1024,
#     temperature=0.2,
# )

llm = LLM(
    model="huggingface/Qwen/Qwen2.5-7B-Instruct",
    api_key=os.environ["HUGGINGFACEHUB_API_TOKEN"],
    temperature=0.2,
    max_tokens=1024,
)
```

```python
search_tool = TavilySearchTool(
    max_results=5
)
```

```python
researcher = Agent(
    role="Senior AI Researcher",

    goal="""
    Determine whether a claim is supported, contradicted, or unsupported by
    available evidence.

    Never stop after finding a single supporting article.

    Gather evidence from multiple independent and reliable sources before
    reaching a conclusion.

    Deliberately search for evidence that both supports and contradicts
    the claim.

    Continue investigating until there is sufficient evidence to justify
    a confident verdict.
    """,

    backstory="""
    You are an experienced investigative journalist and professional
    fact-checker.

    You are skeptical of every claim until it is supported by evidence.

    You never assume that the first search result is correct.

    You actively look for conflicting reports, official statements,
    expert opinions, and multiple independent confirmations.

    You explain your reasoning transparently and clearly separate facts,
    assumptions, and uncertainty.

    If evidence is insufficient, you explicitly state that the claim
    cannot currently be verified rather than guessing.

    """,

    llm=llm,

    tools=[search_tool],

    allow_delegation=False,

    verbose=True
)
```

```python
topic = "Pakistan shot down 6 indian Rafael Jets in their last war interaction. "

research_task = Task(
    description=f"""
    Conduct comprehensive research on the following topic:

    {topic}

    Your objectives are:

    1. Determine what the user is actually asking.
    2. Search multiple reliable sources.
    3. Gather the most recent and relevant information.
    4. Cross-check facts across different sources.
    5. Identify any conflicting information.
    6. Summarize the findings objectively.
    7. Clearly distinguish between confirmed facts and uncertain information.
    8. Include references for every major claim.

    Do not rely on a single source.
    """,

    expected_output="""
    A well-structured markdown report containing:

    - Executive Summary
    - Key Findings
    - Supporting Evidence
    - Conflicting Information (if any)
    - Confidence Assessment
    - References
    """,

    agent=researcher
)
```

```python
crew = Crew(
    agents=[researcher],
    tasks=[research_task],
    verbose=True
)
```

```python
result = await crew.kickoff_async()

print(result)
```
